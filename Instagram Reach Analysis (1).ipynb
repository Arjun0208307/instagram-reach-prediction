```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler , MinMaxScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
#Now Lets Apply Linear Model Witout Scalling 
```

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(X_train_scaled,y_train)
pred=model.predict(X_test_scaled)

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score,root_mean_squared_error
```

print("R_score",r2_score(pred,y_test))
print("Mean_Absolute_Error",mean_absolute_error(pred,y_test))
print("Mean_Squared_Error",mean_absolute_error(pred,y_test))

```python
#Now Lets Apply Random Forest  
from sklearn.ensemble import RandomForestRegressor
```

rf=RandomForestRegressor(n_estimators=50,random_state=42)
model2=rf.fit(X_train_scaled,y_train)
rf_pred=model2.predict(X_test_scaled)

mae = mean_absolute_error(y_test, rf_pred)
mse = mean_squared_error(y_test, rf_pred)
r2 = r2_score(y_test, rf_pred)
rmse = root_mean_squared_error(y_test, rf_pred)
print("R² Score:", r2)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)

```python
#Lets Apply decision tree
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=42)
model3=dt.fit(X_train_scaled,y_train)
dt_pred=model3.predict(X_test_scaled)
```

```python


mae = mean_absolute_error(y_test, dt_pred)
mse = mean_squared_error(y_test, dt_pred)
r2 = r2_score(y_test, dt_pred)
rmse = root_mean_squared_error(y_test, dt_pred)
print("R² Score:", r2)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
```
